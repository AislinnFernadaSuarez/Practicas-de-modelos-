Ejercicio 1: Simular un volado

```python
def volado():
 if rd.random() <=0.5:
  return "cara"
 else:
  return "cruz"
```

```python
volado()
```

Ejercico 2: Use la función anterior para estimar la probabilidad de obtener "cara"

```python
def estimar_probabilidad(n):
  caras = 0
  for _ in range(n):
    if volado() == "cara":
      caras += 1
  return caras / n
```

```python
estimar_probabilidad(1000)
```

```python
import matplotlib.pyplot as plt
n = 1000
caras = 0
probs = []

for i in range(1, n + 1):
    if volado() == "cara":
        caras += 1
    probs.append(caras / i)

plt.plot(probs)
plt.axhline(0.5)
plt.show()
```

Caminatas al azar

```python
import random as rd
import matplotlib.pyplot as plt

def camiale(n, p):
    posicion = 0
    trayectoria = [posicion]

    for i in range(n):
        if rd.random() <= p:
            posicion += 1
        else:
            posicion -= 1
        trayectoria.append(posicion)

    return trayectoria


p = 0.7
n = 100

trayectoria = camiale(n, p)

plt.plot(trayectoria)
plt.xlabel("Paso")
plt.ylabel("Posición")
plt.title("Caminata Aleatoria 1D para n=0.7 y p=100")
plt.axhline(0)
plt.show()
```

Al analizar esta simulación podemos responder:
¿El punto regresa al origen?

Observamos que al inicio puede regresar al origen, pero en la simulación no regresa de forma repetida. Conforme avanzan los pasos, el punto se aleja del cero porque es más probable moverse a la derecha. Esto se debe a que p>q.

Si regresa: ¿En cuánto tiempo regresa pro primera vez? En esta simulación, el punto regresa en los primeros pasos. Por ejemplo, en la gráfica cuando n=100 y p=0.7 se observa que al rededor de los primeros 10-15 pasos la trayectoria cruza el cero. Después de ese punto, el movimiento es en su mayoría a la derecha.

¿Cuánto se aleja, en promedio, del origen? En cada paso, el punto avanza una unidad con probabilidad 0.7 y retrocede una unidad con probabilidad 0.3, por lo que el avance promedio por paso es 0.7-0.3=0.4. Después de 100 pasos, el alejamiento promedio del origen es 100×0.4=40. Por eso, aunque la trayectoria exacta varía por el azar, la simulación muestra valores cercanos a 40.

¿Cuál es su alejamiento máximo? Es el valor más grande que alcanza la posición durante la caminata. En este ejemplo el alejamiento máximo posible es 100 unidades.

¿Este proceso es una cadena de Markov? Sí, es una cadena de Markov ya que en cada paso la nueva posición depende de la actual y de la probabilidad de moverse a la derecha o izquierda.
