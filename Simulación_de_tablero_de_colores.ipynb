Algoritmo para simular una cadena de markov
dado el pseudocódigo visto en clase  

```python
import numpy as np

def simu_mkv(α, P, n):

    m = len(α)  # número de estados

    X = np.zeros(n+1, dtype=int)

    # Generar X0 de acuerdo a α
    X[0] = np.random.choice(m, p=α)

    # Simulación de la cadena
    for i in range(1, n+1):

        j = X[i-1] # estado anterior

        p = P[j]   # tomar el renglón j de P

        X[i] = np.random.choice(m, p=p)   # generar Xi ~ p

    return X
```

Ahora probaremos el código con el problema del tablero de colores

Definimos los parámetros

```python
import matplotlib.pyplot as plt

# distribución inicial uniforme
α = np.array([1/4, 1/4, 1/4, 1/4])

# matriz de transición
P = np.array([
    [1/4, 1/2, 1/4, 0],
    [0, 1/4, 1/2, 1/4],
    [1/4, 0, 1/4, 1/2],
    [1/2, 1/4, 0, 1/4]
])

# número de pasos
n = 100000

# simulación de la cadena
X = simu_mkv(α, P, n)
```

Para analizar el comportamiento a largo plazo, calculamos la frecuencia con la que se visita cada sector

```python
conteo = np.zeros(4)

for estado in X:
    conteo[estado] += 1

prob = conteo/(n+1)

print("Probabilidades aproximadas de cada sector:")
print(prob)
```

Después de simular la cadena durante un número suficientemente grande de pasos en este caso n=100000, podemos observar que las probabilidades estimadas para los cuatro sectores del tablero se aproximan a (0.25,0.25,0.25,0.25). Esto implica que a largo plazo todos los sectores del tablero son igualmente probables, por lo que la distribución estacionaria del proceso es uniforme.
