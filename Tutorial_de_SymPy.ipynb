Tutorial de cómo usar SymPy en python

Primero vamos a importar la libreria necesaria

```python
import sympy as sp
import networkx as nx
import matplotlib.pyplot as plt

sp.init_printing()
```

Definimos variables para poder escribir expresones algebráicas

```python
x, y, z = sp.symbols('x y z')

expr = x**2 + 3*x + 2
expr
```

Usaremos expand() para el producto de dos expresiones

```python
expr = (x + 2)*(x + 3)
sp.expand(expr)
```

Usaremos factor() para la factorización

```python
sp.factor(x**2 + 5*x + 6)
```

Y con ayuda de simplify() vamos a reducir la ecuación

```python
sp.simplify((x**2 - 1)/(x - 1))
```

Operaciones con matrices


```python
A = sp.Matrix([[1,2],
               [3,4]])

B = sp.Matrix([[2,1],
               [0,3]])

A, B
```

Una vez definidas podemos realizar operaciones entre ellas

```python
A*B
```

O bien, el producto en orden inverso y así podemos observar que la multiplicación de matrices no es conmutativa

```python
B*A
```

También podemos hacer la multiplicación por un vector y lo definimos de la siguiente manera:

```python
v = sp.Matrix([x,y])

A*v
```

O bien, se puede usar un vector formado por números

```python
v2 = sp.Matrix([1,2])

A*v2
```

Potencias de matrices

```python
A**2
```

Definimos el determinante de una matriz como

```python
A.det()
```

Y la inversa como

```python
A.inv()
```

Que nos son utiles en sistemas lineales

Solución de sistemas de ecuaciones. Aquí usaremos solve() para devolver los varoles de las variables

```python
x, y = sp.symbols('x y')

eq1 = sp.Eq(2*x + y, 5)
eq2 = sp.Eq(x - y, 1)

sp.solve((eq1,eq2),(x,y))
```

Diagrama de transición

La librería Networkx permite representar grafos dirigidos para modelar diagramas de transición

```python
G = nx.DiGraph()

G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","A")
G.add_edge("A","C")

nx.draw(G, with_labels=True)

plt.show()
```

Diagramas de transición con probabilidades

```python
G = nx.DiGraph()

G.add_weighted_edges_from([
("Soleado","Lluvia",0.3),
("Soleado","Soleado",0.7),
("Lluvia","Soleado",0.5),
("Lluvia","Lluvia",0.5)
])

pos = nx.spring_layout(G)

nx.draw(G,pos,with_labels=True)

labels = nx.get_edge_attributes(G,'weight')

nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

plt.show()
```
