SIMULACIÓN DE VARIABLES ALEATORIAS


Dadas las funciones de disribución de la actividad 4, escribiremos los programas de simulación para generar variables aleatorias, su gráfica; además de verificar la media, varianza y desviación estándar.

Distribución Weibull

```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt


def variable_aleatoria(n, alpha):
    muestra = []
    for i in range(n):

        # Generar U ~ Uniforme(0,1)
        U = rnd.random()

        # Aplicar transformación
        X = (-np.log(U))**(1/alpha)

        muestra.append(X)

    return muestra
```

```python
M1 = variable_aleatoria(10000,2)

print(M1[:10])
```

```python
#Histogrma

plt.hist(M1, bins=30, density=True, color="hotpink")

plt.title("Histograma")
plt.xlabel("x")
plt.ylabel("Frecuencia")
plt.grid()
plt.show()
```

```python
# Ordenar muestra
M1_ordenada = np.sort(M1)

# Valores acumulados
F = np.arange(1,len(M1)+1)/len(M1)

# Graficar
plt.plot(M1_ordenada,F,color="hotpink")

plt.title("Distribución empírica")
plt.xlabel("x")
plt.ylabel("F(x)")

plt.grid()

plt.show()
```

```python
# Media, varianza y desviación estándar

media = np.mean(M1)

varianza = np.var(M1)

desviacion = np.std(M1)

print("Media =", media)

print("Varianza =", varianza)

print("Desviación estándar =", desviacion)
```

La gráfica obtenida refleja el comportamiento típico de la distribución Weibull, ya que la mayor concentración de datos se encuentra en valores pequeños y la frecuencia disminuye conforme los valores aumentan. Esto ocurre debido a la forma de su función de densidad para el parámetro alpha utilizado.

Distrubución Gumbel

```python
import numpy as np
import matplotlib.pyplot as plt

def variable_aleatoria(n):

    muestra = []

    for i in range(n):

        # Generar U ~ Uniforme(0,1)
        U = np.random.random()

        # Aplicar transformación
        X = -np.log(-np.log(U))

        # Guardar dato
        muestra.append(X)

    return muestra
```

```python
M1 = variable_aleatoria(10000)

print(M1[:10])
```

```python
# Histograma

plt.hist(M1,
         bins=30,
         density=True,
         color="yellow")

plt.title("Histograma")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.grid(color="yellow")

plt.show()
```

```python
# Media, varianza y desviación estándar
# Ordenar muestra
M1_ordenada = np.sort(M1)

# Probabilidades acumuladas
F = np.arange(1,len(M1)+1)/len(M1)

# Graficar
plt.plot(M1_ordenada,
         F,
         color="yellow",
         linewidth=3)

plt.title("Distribución empírica")
plt.xlabel("x")
plt.ylabel("F(x)")

plt.grid(color="yellow")

plt.show()
```

```python
# Media, varianza y desviación estándar

media = np.mean(M1)

varianza = np.var(M1)

desviacion = np.std(M1)

print("Media =", media)

print("Varianza =", varianza)

print("Desviación estándar =", desviacion)
```

La gráfica muestra que la mayoría de los datos se concentran en ciertos valores, mientras que los valores muy grandes aparecen con menos frecuencia. Este comportamiento es típico de la distribución Gumbel, ya que se usa para modelar eventos extremos.

Distribución Cauchy

```python
import numpy as np
import matplotlib.pyplot as plt

def variable_aleatoria(n):

    muestra = []

    for i in range(n):

        # Generar U ~ Uniforme(0,1)
        U = np.random.random()

        # Aplicar transformación inversa
        X = np.tan(np.pi*(U - 1/2))

        muestra.append(X)

    return muestra
```

```python
M1 = variable_aleatoria(10000)

print(M1[:10])
```

```python
# Histograma

plt.hist(M1,
         bins=80,
         density=True,
         color="green",
         range=(-10,10))

plt.title("Histograma")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.grid(color="green")

plt.show()
```

```python
# Ordenar muestra
M1_ordenada = np.sort(M1)

# Probabilidades acumuladas
F = np.arange(1,len(M1)+1)/len(M1)

# Graficar
plt.plot(M1_ordenada,
         F,
         color="green",
         linewidth=3)

plt.title("Distribución empírica")
plt.xlabel("x")
plt.ylabel("F(x)")

plt.grid(color="green")

plt.xlim(-10,10)

plt.show()
```

```python
# Media, varianza y desviación estándar

media = np.mean(M1)

varianza = np.var(M1)

desviacion = np.std(M1)

print("Media =", media)

print("Varianza =", varianza)

print("Desviación estándar =", desviacion)
```

La gráfica muestra que la mayoría de los datos se concentran cerca del centro, pero también aparecen valores muy extremos con cierta frecuencia, lo cual es característico de la distribución Cauchy debido a sus colas largas.

Esto ocurre debido a la transformación aplicada, la cual genera el comportamiento característico de la distribución logística y permite aproximar adecuadamente su comportamiento teórico.

Distribución Pareto

```python
import numpy as np
import matplotlib.pyplot as plt

def variable_aleatoria(n,alpha):

    muestra = []

    for i in range(n):

        # Generar U ~ Uniforme(0,1)
        U = np.random.random()

        # Aplicar transformación inversa
        X = U**(-1/alpha)

        muestra.append(X)

    return muestra
```

```python
M1 = variable_aleatoria(10000,2)

print(M1[:10])
```

```python
# Histograma

plt.hist(M1,
         bins=80,
         density=True,
         color="blue",
         range=(0,10))

plt.title("Histograma")
plt.xlabel("x")
plt.ylabel("Frecuencia")

plt.grid(color="blue")

plt.show()
```

```python
# Ordenar muestra
M1_ordenada = np.sort(M1)

# Probabilidades acumuladas
F = np.arange(1,len(M1)+1)/len(M1)

# Graficar
plt.plot(M1_ordenada,
         F,
         color="blue",
         linewidth=3)

plt.title("Distribución empírica")
plt.xlabel("x")
plt.ylabel("F(x)")

plt.grid(color="blue")

plt.xlim(0,10)

plt.show()
```

```python
# Media, varianza y desviación estándar


media = np.mean(M1)

varianza = np.var(M1)

desviacion = np.std(M1)

print("Media =", media)

print("Varianza =", varianza)

print("Desviación estándar =", desviacion)
```

La simulación muestra un comportamiento desigual en la distribución de los datos, donde una parte importante de la probabilidad se concentra en pocos valores. Este tipo de comportamiento es característico de la distribución Pareto y permite modelar situaciones reales donde los resultados no se reparten de manera uniforme.
