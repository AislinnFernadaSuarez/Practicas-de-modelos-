Serpientes y Escaleras

Resolveremos analíticamente y por simulación la interrogante: ¿Cúal es el número promedio de tiradas necesarias para terminar el juego de serpientes y escaleras en el tablero?

Para responder esto, notemos que tenemos las siguinetes conexiones:

* Serpiente: **13→4**

* Serpiente: **17→10**

* Escalera: **3→11**

* Escalera: **15→19**

Note que el tablero contiene 20 casillas por lo que la meta es llegar a la casilla 20; además se usara un dado de 6 caras.

Denotamos Ei=el número esperado de tiradas para llegar a la casilla 20 comenzando desde i. Entonces:


```python
import sympy as sp
import numpy as np
import random
import matplotlib.pyplot as plt

seresc = {
    3:11,
    15:19,
    13:4,
    17:10
}

def mover(posicion, dado):
    nueva = posicion + dado

    # Si excede 20, se queda ahí
    if nueva > 20:
        nueva  = posicion

    # Caer en seresc
    if nueva in seresc:
        nueva = seresc[nueva]

    return nueva
```

Construcción matricial

Queremos resolver:

(I−P)E=1

Donde:

P es la matriz de transición entre estados transitorios.
E es el vector de tiempos esperados.
1 es un vector columna de unos.

```python
n = 20

P = sp.zeros(19,19)

for i in range(1,20):

    for dado in range(1,7):

        j = mover(i,dado)

        # Solo estados transitorios
        if j < 20:
            P[i-1,j-1] += sp.Rational(1,6)
```

```python
#Visualizamos la matriz
P
```

Resolver el sistema

La ecuación es:

E=((I−P)^−1)*1




```python
I = sp.eye(19)

unos = sp.ones(19,1)

E = (I - P).inv() * unos
```

```python
for i in range(19):
    print(f"E_{i+1} =", sp.N(E[i],6))
```

```python
promedio_analitico= sp.N(E[0],6)
promedio_analitico
```

Interpretación del resultado analítico

El valor obtenido representa el número esperado de tiradas necesarias para llegar a la casilla final comenzando desde la casilla 1.

Las serpientes aumentan el tiempo esperado porque hacen retroceder al jugador, mientras que las escaleras reducen el número promedio de movimientos necesarios.

El sistema se modela como una cadena de Markov absorbente donde:

La casilla 20 es absorbente.
Las demás casillas son estados transitorios.

Simulación de Monte Carlo

```python
def partida():

    posicion = 1
    tiros = 0

    while posicion != 20:

        dado = random.randint(1,6)

        posicion = mover(posicion,dado)

        tiros += 1

    return tiros
```

```python
N = 100000

resultados = [partida() for _ in range(N)]

promedio_simulado = np.mean(resultados)

promedio_simulado
```

```python
print("Promedio analitico:", promedio_analitico)
print("Promedio simulado:", promedio_simulado)
```

Laberinto del Ratón

Ahora, resolveremos analitica y por simulación, ¿Cuál es la probabilidad de que un ratón, iniciando  en la casilla 0, alcance la comida?

Sabemos que dado el siguiente laberinto:

* La casilla **7** contiene comida (estado de éxito).
* La casilla **8** contiene una descarga eléctrica (estado de fracaso).
* El ratón inicia en la casilla **0**.
* En cada paso el ratón elige uniformemente uno de los movimientos permitidos.

Queremos calcular:

P(alcanzar comida antes que shock∣X0=0)

Para esto, resolveremos el problema de dos formas: analíticamente mediante cadenas de Markov y por simulación Monte Carlo.




Cadena de Markov

Definimos el conjunto de estados:

S={0,1,2,3,4,5,6,7,8},

donde

* 7: absorbente (comida)
* 8: absorbente (shock)

Considerando los movimientos posibles del laberinto:

* 0→{1,2}
* 1→{0,3,7}
* 2→{0,3,8}
* 3→{1,2,4,5}
* 4→{3,6,7}
* 5 →{3,6,8}
* 6→{4,5}

Como el ratón elige aleatoriamente entre las salidas disponibles:

P(i→j)=1/número de vecinos de i

Definimos las ecuaciones de probabilidad:

pi=P(llegar a 7 antes que 8∣X0=i)

Así p7=1 porque ya está en la comida, y p8=0 porque ya cayó en shock.

Sistema de ecuaciones

Usando esperanza condicional:
* p0= 1/2​(p1+p2)
* p1=1/3(p0+p3+1)
* p2=173(p0+p3) porque ir a 8 aporta probabilidad 0.
* p3=1/4(p1+p2​+p4+p5)
* p4=1/3(p3+p6+1)
* p5=1/38p3+p6)
* p6=1/2(p4+p5)

```python
import sympy as sp

p0,p1,p2,p3,p4,p5,p6 = sp.symbols('p0 p1 p2 p3 p4 p5 p6')

eqs = [
    sp.Eq(p0, (p1+p2)/2),
    sp.Eq(p1, (p0+p3+1)/3),
    sp.Eq(p2, (p0+p3)/3),
    sp.Eq(p3, (p1+p2+p4+p5)/4),
    sp.Eq(p4, (p3+p6+1)/3),
    sp.Eq(p5, (p3+p6)/3),
    sp.Eq(p6, (p4+p5)/2)
]

sol = sp.solve(eqs,[p0,p1,p2,p3,p4,p5,p6])
sol
```

```python
A = sp.Matrix([
    [1, -sp.Rational(1,2), -sp.Rational(1,2), 0, 0, 0, 0],
    [-sp.Rational(1,3), 1, 0, -sp.Rational(1,3), 0, 0, 0],
    [-sp.Rational(1,3), 0, 1, -sp.Rational(1,3), 0, 0, 0],
    [0, -sp.Rational(1,4), -sp.Rational(1,4), 1,
     -sp.Rational(1,4), -sp.Rational(1,4), 0],
    [0,0,0,-sp.Rational(1,3),1,0,-sp.Rational(1,3)],
    [0,0,0,-sp.Rational(1,3),0,1,-sp.Rational(1,3)],
    [0,0,0,0,-sp.Rational(1,2),-sp.Rational(1,2),1]
])

b = sp.Matrix([
    0,
    sp.Rational(1,3),
    0,
    0,
    sp.Rational(1,3),
    0,
    0
])
```

```python
p = A.LUsolve(b)
p
```

Ahora verificaremos el resultado mediante simulación de Monte Carlo.

```python
import random

moves = {
    0:[1,2],
    1:[0,3,7],
    2:[0,3,8],
    3:[1,2,4,5],
    4:[3,6,7],
    5:[3,6,8],
    6:[4,5]
}

def trial():
    state = 0

    while state not in [7,8]:
        state = random.choice(moves[state])

    return 1 if state == 7 else 0

```

```python
N = 100000

results = [trial() for _ in range(N)]

prob = sum(results)/N

print(prob)
```

Modelando el movimiento del ratón como una cadena de Markov absorbente se obtuvo el sistema de ecuaciones asociado a las probabilidades de absorción.

Al resolver el sistema analíticamente con SymPy y verificar mediante simulación Monte Carlo, se concluye que:

P(alcanzar la comida antes del shock∣X0=0)=1/2

La simulación confirma el resultado teórico, mostrando cómo las cadenas de Markov permiten modelar procesos aleatorios de decisión y absorción de manera exacta.
